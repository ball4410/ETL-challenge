```python
#Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Store csv to Data Frame
csv_file = "Resources/degrees-that-pay-back.csv"
degrees_df = pd.read_csv(csv_file)
degrees_df

new_degrees_df = degrees_df[["Undergraduate Major", "Mid-Career Median Salary",
                            "Mid-Career 25th Percentile Salary", 
                             "Mid-Career 75th Percentile Salary"]].copy()
new_degrees_df = new_degrees_df.rename(columns = {"Undergraduate Major": "major",
                                "Mid-Career Median Salary": "median_salary",
                                "Mid-Career 25th Percentile Salary": "first_percentile",
                                "Mid-Career 75th Percentile Salary": "third_percentile"})

new_degrees_df.head()
```

```python
#Clean Data
new_degrees_df['median_salary'] = [x.strip('$') for x in new_degrees_df['median_salary']]
new_degrees_df['first_percentile'] = [x.strip('$') for x in new_degrees_df['first_percentile']]
new_degrees_df['third_percentile'] = [x.strip('$') for x in new_degrees_df['third_percentile']]

new_degrees_df['median_salary']=new_degrees_df['median_salary'].str.replace(',','')
new_degrees_df['first_percentile']=new_degrees_df['first_percentile'].str.replace(',','')
new_degrees_df['third_percentile']=new_degrees_df['third_percentile'].str.replace(',','')
new_degrees_df.head()
```

```python
#Store csv to Data Frame
csv_file2 = "Resources/salaries-by-college-type.csv"
college_type_df = pd.read_csv(csv_file2)
college_type_df

new_types_df = college_type_df[["School Name", "School Type",
                             "Mid-Career Median Salary",
                            "Mid-Career 25th Percentile Salary", 
                             "Mid-Career 75th Percentile Salary"]].copy()

new_types_df = new_types_df.rename(columns = {"School Name": "school_name",
                               "School Type": "school_type",
                                "Mid-Career Median Salary": "median_salary",
                                "Mid-Career 25th Percentile Salary": "first_percentile",
                                "Mid-Career 75th Percentile Salary": "third_percentile"})

new_types_df.head()
```

```python
#Clean Data
new_types_df['median_salary'] = [x.strip('$') for x in new_types_df['median_salary']]
new_types_df['first_percentile'] = [x.strip('$') for x in new_types_df['first_percentile']]
new_types_df['third_percentile'] = [x.strip('$') for x in new_types_df['third_percentile']]

new_types_df['median_salary']=new_types_df['median_salary'].str.replace(',','')
new_types_df['first_percentile']=new_types_df['first_percentile'].str.replace(',','')
new_types_df['third_percentile']=new_types_df['third_percentile'].str.replace(',','')
new_types_df.head()
```

```python
#Store csv to Data Frame
csv_file3 = "Resources/salaries-by-region.csv"
region_df = pd.read_csv(csv_file3)
region_df

new_regions_df = region_df[["School Name", "Region",
                             "Mid-Career Median Salary",
                            "Mid-Career 25th Percentile Salary", 
                             "Mid-Career 75th Percentile Salary"]].copy()
new_regions_df = new_regions_df.rename(columns = {"School Name": "school_name",
                                "Region": "region",
                                "Mid-Career Median Salary": "median_salary",
                                "Mid-Career 25th Percentile Salary": "first_percentile",
                                "Mid-Career 75th Percentile Salary": "third_percentile"})
new_regions_df.head()
```

```python
#Clean Data
new_regions_df['median_salary'] = [x.strip('$') for x in new_regions_df['median_salary']]
new_regions_df['first_percentile'] = [x.strip('$') for x in new_regions_df['first_percentile']]
new_regions_df['third_percentile'] = [x.strip('$') for x in new_regions_df['third_percentile']]

new_regions_df['median_salary']=new_regions_df['median_salary'].str.replace(',','')
new_regions_df['first_percentile']=new_regions_df['first_percentile'].str.replace(',','')
new_regions_df['third_percentile']=new_regions_df['third_percentile'].str.replace(',','')
new_regions_df.head()
```

```python
#Store JSON data into DataFrame
json_file = "Resources/schoolInfo.json"
school_info_df = pd.read_json(json_file)
school_info_df
```

```python
#Clean DataFrame
new_school_df = school_info_df[["displayName", "city", "state",  "overallRank", "enrollment", 
                               "acceptance-rate", "tuition", "percent-receiving-aid", "cost-after-aid",
                                "hs-gpa-avg", "act-avg", "sat-avg"]].copy()
new_school_df = new_school_df.rename(columns = {"displayName": "school_name",
                               "overallRank": "rank",
                               "acceptance-rate": "acceptance_rate",
                               "percent-receiving-aid": "percent_receiving_aid",
                               "cost-after-aid": "cost_after_aid",
                               "hs-gpa-avg": "avg_hs_gpa",
                               "act-avg": "avg_act_score", "sat-avg": "avg_sat_score"})
```

CREATE TABLE degrees (
	major VARCHAR, 
	median_salary FLOAT, 
	first_percentile FLOAT,
	third_percentile FLOAT
);

CREATE TABLE college_types (
	school_name VARCHAR, 
	school_type VARCHAR, 
	median_salary FLOAT, 
	first_percentile FLOAT,
	third_percentile FLOAT
);

CREATE TABLE regions (
	school_name VARCHAR, 
	region VARCHAR, 
	median_salary FLOAT, 
	first_percentile FLOAT,
	third_percentile FLOAT
);

CREATE TABLE school_info (
	school_name VARCHAR, 
	city VARCHAR, 
	state VARCHAR(2),
	rank INT,
	enrollment INT,
	acceptance_rate INT,
	tuition INT, 
	percent_receiving_aid FLOAT, 
	cost_after_aid FLOAT,
	avg_hs_gpa FLOAT,
	avg_act_score INT,
	avg_sat_score INT
);

```python
#Load data using pandas
new_degrees_df.to_sql(name="degrees", con=engine, if_exists='append', index=False)
new_types_df.to_sql(name = "college_types", con=engine, if_exists='append', index=False)
```
